## Class and Object

```python
class Parrot:

    # class attribute
    species = "bird"

    # instance attribute
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

species라고 직접 attribute를 선언하는 경우도 있지만, __init__에서 self.name , self.age와 같이 attribute를 선언할 수도 있다 !

```python
# instantiate the Parrot class
blu = Parrot("Blu", 10) # 첫번째 인자 = name, 두번째 인자 = age
woo = Parrot("Woo", 15) # 첫번째 인자 = name, 두번째 인자 = age

# access the class attributes
print("Blu is a {}".format(blu.__class__.species))
print("Woo is also a {}".format(woo.__class__.species))

# access the instance attributes
print("{} is {} years old".format( blu.name, blu.age))
print("{} is {} years old".format( woo.name, woo.age))
```

```python
class Parrot:
    
    # instance attributes
    def __init__(self, name, age): # special method 또는 magic method , self는 관행과 권장사항일 뿐 다른거로 바꿔도 된다. this 같은 것도 된다.
        self.name = name
        self.age = age
    
    # instance method
    def sing(self, song): # self가 들어가면 instance method
        return "{} sings {}".format(self.name, song)

    def dance(self): # self가 들어가면 instance method
        return "{} is now dancing".format(self.name)

# instantiate the object
blu = Parrot("Blu", 10)

# call our instance methods
print(blu.sing("'Happy'"))
print(blu.dance())
```

```python
# child class
class Penguin(Bird): # <= 상속을 하는 방법 펭귄은 Bird를 상속받았으므로 Bird의 모든 메소드를 사용할 수 있다.
    
    def __init__(self):
        # call super() function
        super(Penguin,self).__init__() # SUPER ! => 부모의 것을 사용하고 자신의 __init__ 의 내용을 그 다음에 실행함
        print("Penguin is ready")

    def whoisThis(self): # 오버라이딩
        print("Penguin")

    def swim(self): # 오버 라이딩
        print("Can't swim")
        
    def run(self): # 새로운 메소드 추가
        print("Run faster")
```

```python
peggy = Penguin()
peggy.whoisThis()
peggy.swim()
peggy.run()
```

super()에서 인자를 다양하게 넣을 수 있다는 것은 다중상속을 의미한다.
- super(Bird, self) 라고 하는 것은 super() 라고 하는 것과 똑같다.
- super(super를 받고싶은 XXX 클래스, self) 이런 식으로 지정하는 것이 가능하다.

## Encapsulation

파이썬은 캡슐화를 제대로 지원하지 않는다.

내부 정보를 잘 간직하기 위해서 사용하는데, 파이썬에서는 mutable을 지원하기 때문에 Java만큼 제대로 사용되지는 않는다. '__변수' 로 private 비슷하게 사용할 수는 있으나, 이 마저도 외부에서 접근이 가능하다.

클래스 자체에서 캡슐화 하면 좋겠지만, private 같은 개념이 없기 때문에 이것도 제대로 캡슐화 하기 어렵다.

```python
class Computer:

    def __init__(self):
        self.__maxprice = 900

    def sell(self):
        print("Selling Price: {}".format(self.__maxprice))

    def setMaxPrice(self, price):
        self.__maxprice = price

c = Computer()
c.sell()

# change the price
c.__maxprice = 1000
c.sell()

# using setter function
c.setMaxPrice(1000)
c.sell()
```

## Int 클래스를 상속받아서 나만의 Int 클래스를 만들기

```python
class MyINT(int) :
    def __init__(self, num) :
        self.num = num
    def __add__(self, other) :
        print('문근영')
        return self.num + other + 200
    def __repr__(self) :
        return '안 알려줌'
    def __str__(self) :
        return 'Simon'
    def __sub__(self, other) :
        print('조작된 __sub__로 변형된 값이 나올 것이다. >>> ', end=" ")
        return self.num - other + 500
    def __radd__(self, other) : # 오른쪽에 올 때 radd ....(?)
        return self.num + other + 1
```

self 자체는 object이지 값 자체가 아니다. 따라서 self.num = num 이런 식으로 !!!
