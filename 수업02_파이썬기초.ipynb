### 예외처리 

```python
a = input()

try :
    b = int(a)
except :
    print('숫자 입력해 !')
else : # 예외처리에서의 else문의 기능 : 예외가 발생하지 않았을 경우 처리
    print('잘 입력하셨어요')
finally :
    print('무조건 실행')
```

```python
while(True) :

    try : 
        tmp = int(input('숫자 : '))
        num = int(tmp) if tmp > 1 and tmp < 10 else int("a")
        
    except :
        print('다시 입력하시오.')
        print()

    else :
        print()
        print('{}단 출력'.format(num))
        for i in range(1,10) :
            print("{} X {} = {}".format(num,i,num*i))
        break
```
